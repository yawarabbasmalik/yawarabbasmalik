```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('/content/covid_19_indonesia_time_series_all.csv')
df.head()
```

```python
df.info()
```

```python
df['Month'] = df['Date'].apply(lambda x:datetime.datetime.strptime(x, '%m/%d/%Y').strftime('%Y-%m'))
df.head()
```

```python
df.info()
```

```python
df = df[df['Location'] != 'Indonesia']
```

# Exploratory Data Analysis

See the TOP 5 cases of COVID 19 in Indonesia

```python
top_cases = (df.groupby('Location')['New Cases'].sum().reset_index().sort_values(by='New Cases', ascending=False).head(5))
top_cases
```

```python
import plotly.express as px

data = top_cases

plot_top = px.data.gapminder().query("country == 'Indonesia'")
fig = px.bar(plot_top, x=top_cases["Location"], y=top_cases['New Cases'])

fig.show()
```

Comparison of COVID-19 cases in the Top 5 Provinces in 2021

```python
cov_2022 = df[(df['Month']>='2021-01')]
cov_2022

cov_2022['top_case'] = cov_2022['Location'].apply(lambda x: x if(x in top_cases['Location'].to_list()) else 'other')
cov_2022
```

```python
cov_2022.groupby(['Month', 'top_case'])['New Cases'].sum().unstack().plot()
plt.title('Positive Cases of COVID-19 in 2021\n by Province with the Most Cases', loc='center', pad=20, fontsize=22, color='navy')
plt.xlabel('Date', fontsize=15)
plt.ylabel('Number of Cases', fontsize=15)
plt.grid(color='darkgray', linestyle='dotted', linewidth=0.5)
plt.ylim(ymin=0)
plt.xticks(rotation=45)
plt.legend(loc='right', bbox_to_anchor=(1.35, 0.5), ncol=1, title='Provinces')
plt.show()
```
